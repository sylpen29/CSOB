```python
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
vente_EU = df[["Year", "Platform", "EU_Sales"]]
vente_EU["Year"].astype("int")
Xbox = vente_EU[(vente_EU["Platform"] == "X360")|(vente_EU["Platform"] == "XOne")].groupby("Year").sum()
```

```python
PS = vente_EU[(vente_EU["Platform"] == "PS3")|(vente_EU["Platform"] == "PS4")].groupby("Year").sum()
```

```python
Wii = vente_EU[(vente_EU["Platform"] == "Wii")|(vente_EU["Platform"] == "WiiU")].groupby("Year").sum()
```

```python
plt.figure(figsize=(10,8))
plt.bar(PS.index, PS.EU_Sales, label="Playstation (3 & 4)")
plt.bar(Xbox.index, Xbox.EU_Sales, color="green", label="Xbox (360 & One)")
plt.bar(Wii.index, Wii.EU_Sales, color="red", label="Wii & Wii U", alpha=0.65)
plt.title("Nombre de vente EU pour les 3 grands groupe depuis 2006")
plt.xlabel("Annee")
plt.ylabel("Nombre de vente (million)")
plt.legend()
plt.show()
```
